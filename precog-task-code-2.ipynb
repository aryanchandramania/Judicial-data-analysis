```python
def filt(x):
    vals = acts[acts['ddl_case_id'] == x]["bailable_ipc"].values
    if not len(vals):
        return False
    return vals[0] == 'bailable'
    
```

```python
joineindex = main_df.merge(acts, on='ddl_case_id', how='inner')
```

```python
joineindex.info()
```

```python
joineindex.date_of_filing = pd.to_datetime(joineindex['date_of_filing'], errors='coerce')
joineindex.info()
```

```python
# joineindex.date_of_decision.replace(r'','2021-12-31')
joineindex.date_of_decision[np.isnan(joineindex.date_of_decision)] = '2021-12-31'
```
